```python
import matplotlib.pyplot as plt
import numpy as np
from functionsUtils import load_results
#
plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
```

```python
# Data
alternatives = ['Q-random', 'Best individual APs', 'CPU-local ES', 'CPU-local SG','(Succ.) CPU-local SG','(Succ.) CPU-local ES', 'Sequential Greedy']
values = [39.7576, 51.0572, 53.1731, 53.1831,  53.8660,  54.1088, 54.1757]  # Replace with your values

fig, ax = plt.subplots()
plt.grid(linestyle='--', alpha=0.5)

spacing = 10 # Spacing factor
x_positions = np.arange(len(alternatives)) * spacing
# Create the bar chart
plt.bar(x_positions, values, width=8, color=['black', 'green', 'skyblue', 'lightcoral', 'red', 'blue', 'firebrick'])

plt.xticks(x_positions, alternatives, rotation=70, fontsize=12)
plt.ylim([35, 58])

# Add labels and title
plt.ylabel('Average sum-rate [bits/s/Hz]', fontsize=12)

# Annotate values on bars
for i, value in enumerate(values):
    plt.text(x_positions[i], value + 1, str(value), ha='center', fontsize=10)

# Show the plot
plt.tight_layout()
plt.draw()
plt.show()

image_format = 'png' # e.g .png, .svg, etc.
image_name = 'AveSE_BarGraph.png'
fig.savefig(image_name, format=image_format, dpi=400)

```

```python
import math

T = 8
# Q = 4

combinaciones = [math.comb(T, q) for q in range(1, T+1)]
print(combinaciones)

secuencias = [(T - (T-q) + 1) * ((T-q) + T) // 2 for q in range(0, T)]
print(secuencias)

# Plot the results
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Plot combinations
ax[0].bar(range(1, T+1), combinaciones, color='blue')
ax[0].set_title('Combinations')
ax[0].set_xlabel('q')
ax[0].set_ylabel('Combinations')
ax[0].set_ylim([0, 80])

# Plot sequences
ax[1].bar(range(0, T), secuencias, color='green')
ax[1].set_title('Sequences')
ax[1].set_xlabel('q')
ax[1].set_ylabel('Sequences')
ax[1].set_ylim([0, 80])


# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
