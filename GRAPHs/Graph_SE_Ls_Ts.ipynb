```python
import matplotlib.pyplot as plt
import numpy as np
from functionsUtils import load_results

plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
```

```python
# Load results
results = load_results('./VARIABLES_SAVED/SE_Ls_Ts_NbrSetps_50_K_100.pkl')
bestgains = results['bestgains_individualAPs']
Q_random = results['Q_random']
GNN = results['GNN']
successive_local_ES = results['local_SG']

# Tuples for x-axis labels
configurations = [(65, 4), (85, 5), (100, 6), (115, 7), (130, 8), (145, 9)]
x_positions = np.arange(len(configurations))  # Numeric positions for the x-axis

# Plot
fig = plt.figure()
ax = fig.add_subplot(111)

plt.grid(visible=True, linestyle='--')
ax.set_ylabel(r'sum-SE [bit/s/Hz]', size=15)
plt.xlabel(r'(\textit{L}, \textit{T})', size=15)
plt.plot(x_positions, GNN, label="Pixel-based CF-mMIMO", color='g')
plt.plot(x_positions, successive_local_ES, label="SLES", color='b')
plt.plot(x_positions, Q_random, label=r"$W_\textsubscript{max}$-random", color='m')
plt.plot(x_positions, bestgains, label="MG", color='c')

# Set x-axis ticks and labels
plt.xticks(x_positions, configurations, rotation=30, fontsize=10)

plt.ylim(152,225)
plt.xlim(-0.1, len(configurations) - 0.9)
# plt.xticks(np.arange(50, 200, step=25))

plt.legend(fontsize=10)

plt.draw()
plt.show()

# Save the figure
image_format = 'pdf'  # e.g .png, .svg, etc.
image_name = 'SE_Ls_Ts.pdf'
fig.savefig(image_name, format=image_format, dpi=400, bbox_inches='tight')
```
