```python
import matplotlib.pyplot as plt
import numpy as np
from functionsUtils import load_results

plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
```

```python
# results_400_1 = load_results('./VARIABLES_SAVED/SE_CDF_K_100_NbrSetps_5_L_400_N_1.pkl')
results_100_4 = load_results('./VARIABLES_SAVED/SE_CDF_K_100_NbrSetps_10_L_100_N_4.pkl')

bestgains_L_100 = results_100_4['bestgains_individualAPs'][0, :]
# bestgains_L_400 = results_400_1['bestgains_individualAPs'][1, :]

Q_random_L_100 = results_100_4['Q_random'][0, :]
# Q_random_L_400 = results_400_1['Q_random'][1, :]

GNN_L_100 = results_100_4['GNN'][0, :]
# GNN_L_400 = results_400_1['GNN'][1, :]

successive_local_ES_L_100 = results_100_4['successive_local_ES'][0, :]
# successive_local_ES_L_400 = results_400_1['successive_local_ES'][1, :]


fig, ax = plt.subplots()
plt.grid(visible = True, linestyle='--')

plt.ecdf(bestgains_L_100, label="MG", color='c')
plt.ecdf(Q_random_L_100, label=r"$W_\textsubscript{max}$-random", color='m')
plt.ecdf(successive_local_ES_L_100, label="SLES", color='b')
plt.ecdf(GNN_L_100, label="Pixel-based CF-mMIMO", color='g')

plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, 5.10, step=0.5))

plt.xlim(0, 4.8)
plt.ylim(0, 1.02)

plt.xlabel(r'SE$_{k}$ [bit/s/Hz]', size=15)
plt.ylabel(r'CDF', size=15)

plt.legend(fontsize=10)

plt.draw()
plt.show()

image_format = 'pdf' # e.g .png, .svg, etc.
image_name = 'SEs_CDF.pdf'
fig.savefig(image_name, format=image_format, dpi=400, bbox_inches='tight')
```
